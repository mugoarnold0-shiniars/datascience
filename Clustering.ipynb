```python
import pandas
data = pandas.read_csv("datasets/Mall_Customers.csv")
data
```

```python
data.isnull().sum()
```

```python
# Drop the null values idedified
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
array=data.values
X=array[:,2:5]
X
```

```python
# we use kmeans since we are dealing with continuos data
from sklearn.cluster import KMeans
model = KMeans(n_clusters=15,random_state=42)
model.fit(X)
```

```python
# Let see the clustered created
means = model.cluster_centers_
means
```

```python
# Lets label our clusters
clusters= pandas.DataFrame(means,columns=["Age","Annual Income","Spending Score"])
clusters
```

```python
# Let see who is in group5
data["Cluster_number"] = model.labels_ #get every record group  number
cluster5 = data[data["Cluster_number"]==5 ]
cluster5
```

```python
# exporting data members of group 5 to csv
data[data["Cluster_number"]==5].to_csv('Cluster5.csv')
```
