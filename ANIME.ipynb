```python
import pandas 


data= pandas.read_csv("datasets/anime.csv")
data
```

```python
data.describe()
```

```python
data.groupby("genre").size()
```

```python
data["type"].value_counts().plot.pie(autopct='%1f%%')
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

sns.countplot(x="type",data=data,palette="summer")
plt.title("Type of Anime aganist Total Count")
```

```python
sns.barplot(x="type",y="rating",data=data,palette="Blues")
plt.title("Total Ratings of anime aganist type")
```

```python
sns. scatterplot(x="rating",y="members" , data=data,palette="summer", hue="type")
```

```python
# below is a line graph showing the total sales made over time

data['rating']=pandas.to_numeric(data["rating"])

daily_sales = data.groupby("type")["members"].sum()

# plot
plt.figure(figsize=(10,5))
plt.plot(daily_sales.index,daily_sales.values,marker='o',linestyle="-",linewidth=2)
plt.grid
plt.title("Total Members Over Time")
plt.tight_layout
plt.xlabel("ratings")
plt.ylabel("Members")
```

```python

numeric_cols = ["members","anime_id","rating"]


corr = data[numeric_cols].corr()


plt.figure(figsize=(10,5))
sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Between Numerical variables")
plt.tight_layout
```

```python
payment_counts  = data["type"].value_counts()
plt.figure(figsize=(6,6))
wedges,texts,autotext = plt.pie(
  payment_counts,
  labels=payment_counts .index,
  autopct= '%1.1f%%',
  startangle=90,
  wedgeprops={'width':0.4}

)

center_circle = plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(center_circle)


plt.title("Distribution of Types Of Anime",fontsize=14)
plt.tight_layout()
plt.show()
```
