```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Gender",data=data , palette="Greens" , hue="Property_Area")
plt.title("Distribution of funds In the Areas ")
plt.xlabel("Gender")
plt.ylabel("Loans")

```

```python
sns.barplot(x="Gender",y="ApplicantIncome",data=data,palette="Blues",hue="Property_Area")
plt.title("Total Distribution of Gender aganist Total Income across the Areas")
```

```python
sns.swarmplot(x="Gender",y="LoanAmount",data=data,palette="autumn" , hue="Loan_Status")
```

```python
payment_counts  = data["Property_Area"].value_counts()
plt.figure(figsize=(6,6))
wedges,texts,autotext = plt.pie(
  payment_counts,
  labels=payment_counts .index,
  autopct= '%1.1f%%',
  startangle=90,
  wedgeprops={'width':0.4}

)

center_circle = plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(center_circle)


plt.title("Distribution of Loan Method",fontsize=14)
plt.tight_layout()
plt.show()
```
