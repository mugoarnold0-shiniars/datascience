```python
import pandas as pd
data = pd.read_csv('datasets/Cluster5.csv')
data
```

```python
data.describe()
```

```python
data["Gender"].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Gender",data=data ,palette="Reds"  )
plt.title("Distribution of Males and Females in different Clusters")
plt.xlabel("Different gender")
plt.ylabel("Distribution count")
```

```python
sns.barplot(x="Gender",y="Spending Score",data=data,palette="Blues")
plt.title("Total Distribution of Gender aganist Spending Score")
```

```python
sns.swarmplot(x="Gender",y="Spending Score",data=data,palette="autumn" , hue="Age")
```
