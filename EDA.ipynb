```python
# plottinf graphs based on data we have 
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="gender",data=data ,palette="Reds" , hue="city")
plt.title("Distribution of Males and Females in different Cities")
plt.xlabel("Different gender")
plt.ylabel("Distribution count")
```

```python
# Below is a barplot
# a barplot is a bivariate/multivartible meaning it expects you to pass more than one column
sns.barplot(x="gender",y="total",data=data,palette="Blues",hue="city")
plt.title("Total Distribution of Gender aganist Total purchase")
```

```python
# Below is a swamPlot
sns.swarmplot(x="gender",y="total",data=data,palette="autumn" , hue="payment")
```

```python
# Below is a scatterplot which is an opposite of swarmplot 
sns. scatterplot(x="rating",y="total" , data=data,palette="summer", hue="gender")
```

```python
# Below is a lie graph 
import matplotlib.dates as mdates
import pandas as pd
import matplotlib.pyplot as plt
# Make sure your date column is datetime
data["date"] = pd.to_datetime(data["date"])

# format the axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())


# crate the graph

sns.lineplot(x="date" , y="total", data=data)

# Force full month range (Janâ€“Dec of that year)
year = data["date"].dt.year.min()  # pick the year of your data
plt.gca().set_xlim(pd.Timestamp(f"{year}-01-01"), pd.Timestamp(f"{year}-12-31"))

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# below is a line graph showing the total sales made over time

data['date']=pd.to_datetime(data["date"])

daily_sales = data.groupby("date")["total"].sum()

# plot
plt.figure(figsize=(10,5))
plt.plot(daily_sales.index,daily_sales.values,marker='o',linestyle="-",linewidth=2)
plt.grid
plt.title("Total Sales Over Time")
plt.tight_layout
plt.xlabel("Date")
plt.ylabel("Total Sales")
```

```python
# Below is a heatmap that enables us to see the correlation btwn different entities

numeric_cols = ["itemprice","quantity","total","customersalary","rating"]

# below we compute the correlation matrix
corr = data[numeric_cols].corr()

# Plot the heatmap

plt.figure(figsize=(10,5))
sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Between Numerical variables")
plt.tight_layout
```
