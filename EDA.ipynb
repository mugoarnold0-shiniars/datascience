```python
# plottinf graphs based on data we have 
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="gender",data=data ,palette="Reds" , hue="city")
plt.title("Distribution of Males and Females in different Cities")
plt.xlabel("Different gender")
plt.ylabel("Distribution count")
```

```python
# Below is a barplot
# a barplot is a bivariate/multivartible meaning it expects you to pass more than one column
sns.barplot(x="gender",y="total",data=data,palette="Blues",hue="city")
plt.title("Total Distribution of Gender aganist Total purchase")
```

```python
# Below is a swamPlot
sns.swarmplot(x="gender",y="total",data=data,palette="autumn" , hue="payment")
```

```python
# Below is a scatterplot which is an opposite of swarmplot 
sns. scatterplot(x="rating",y="total" , data=data,palette="summer", hue="gender")
```

```python
# Below is a lie graph 
import matplotlib.dates as mdates

# format the axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# crate the graph

sns.lineplot(x="date" , y="total", data=data)
```
